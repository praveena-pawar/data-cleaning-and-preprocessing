```python
cleaned_data = df4.copy()
cleaned_data.fillna(0, inplace=True)
```

```python
cleaned_data
```

```python
cleaned_data.fillna(0, inplace=True)
```

```python
cleaned_data
```

```python
age_mean = cleaned_data["age"].mean()
cleaned_data["age"].fillna(age_mean, inplace=True)
```

```python
cleaned_data
```

```python
income_mean = cleaned_data["income"].mean()
cleaned_data["income"].fillna(income_mean)
```

```python
cleaned_data
```

```python
cleaned_data
```

```python
cleaned_data["name"].replace(0, "Unknown", inplace=True)
cleaned_data["name"].fillna("Unknown", inplace=True)
```

```python
cleaned_data
```

```python
cleaned_data["country"].replace(0, "Unknown", inplace=True)
cleaned_data["country"].fillna("Unknown", inplace=True)
```

```python
cleaned_data["gender"].replace(0, "Unknown", inplace=True)
cleaned_data["gender"].fillna("Unknown", inplace=True)
```

```python
cleaned_data
```

```python
mean_income = cleaned_data[cleaned_data["income"] != 0]["income"].mean()
cleaned_data["income"].replace(0, mean_income, inplace=True)
```
